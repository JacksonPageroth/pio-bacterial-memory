```python
#Imports
import numpy as np
import pandas as pd


import matplotlib.pyplot as plt
%matplotlib inline


import seaborn as sns
```

```python
#Dataframe upload
df = pd.read_excel('/Users/jackson/Downloads/export_11_08_2024_20241110093542/11_08_2024-pioreactor_unit_activity_data-worker2-20241110093543.csv')

# print(df.head(5))
# print(df.dtypes)

name_dict = {
    'Worker 2': 'W2 - 35 Control',
    'Worker 3': 'W3 - Replicate 1',
    'Worker 4': 'W4 - Replicate 2',
    'Worker 5': 'W5 - 55 Control'
}

# Replace values in the 'Pioreactor name' column using the dictionary
df['Pioreactor name'] = df['Pioreactor name'].replace(name_dict)

#Plotting the growth curves
f, ax = plt.subplots(figsize=(12, 6.5))
sns.scatterplot(x="Date", y="Real OD",
                hue="Pioreactor name",
                palette='tab10',
                sizes=(1, 8), linewidth=0,
                data=df, ax=ax)
plt.legend()
plt.axhline(y=0.24, color='r', linestyle='--')
ax.set_title('Pioreactor OD Curves')
```
